**Import Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Pandas is used to load and manage the dataset. Seaborn and Matplotlib are used to create visualizations and plots for better understanding of the data

**Load Dataset**

```python
df=pd.read_csv('/content/drive/MyDrive/IRIS.csv')
df.head()
```

Iris dataset is loaded from a CSV file using pandas. The head() function is used to display the first five rows of the dataset so we can see how the data looks and understand its structure

**Check Shape**

```python
df.shape
```

Shape function is used to check the size of the dataset. It tells us how many rows and columns are present in the dataset, which helps us understand how much data we are working with

**Display Column names**

```python
df.columns
```

Columns function is used to display all column names in the dataset. It helps to identify the features and the target variable

**Give information about Dataset**

```python
df.info()
```

info() function provides detailed information about the dataset, including the data types of each column and the number of non-null values. It also helps in checking if there are any missing values in the dataset

**Statistics Summary**

```python
df.describe()
```

describe() function generates summary statistics such as mean, minimum value, maximum value, and standard deviation for all numerical columns. This helps in understanding the overall distribution of the data

**Scatter Plot**

```python
sns.scatterplot(
    x ='sepal_length',
    y ='petal_length',
    hue ='species',
    data = df
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

Scatter plot is created to show the relationship between sepal length and petal length. Different colors are used to represent different species. This visualization helps to see how well the species are separated based on these features

**Histogram**

```python
df.hist(figsize=(10, 6))
plt.show()
```

Histograms are used to show the distribution of each numerical feature in the dataset.It helps to understand how the values are spread and whether the data is normally distributed or skewed

**Box Plot **

```python
sns.boxplot(data = df.drop(columns = 'species'))
plt.title("Box Plot of Iris Features")
plt.show()
```

Box plot is created to visualize the spread of numerical features and to detect any outliers
